```python
FPS = 100
```

```python
BEAT_PATH = '/Users/matthewdavies/Desktop/'
```

```python
example = 'cuidado_HideIt'

example_audio = '%s.flac' % (BEAT_PATH, example)
example_annotation = '%s.beats.txt' %  (BEAT_PATH, example)
annotations = madmom.io.load_onsets(example_annotation)
```

```python
annotations = madmom.io.load_onsets('/Users/matthewdavies/Desktop/cuidado_HideIt.beats.txt')
signal = madmom.audio.Signal('/Users/matthewdavies/Desktop/cuidado_HideIt.flac')
```

```python
plt.plot(signal)
plt.xlabel('sample')
plt.ylabel('amplitude')
for ann in annotations:
    plt.axvline(x=ann * signal.sample_rate, color='r', linestyle=':', linewidth=2)
```

```python
spec = madmom.audio.Spectrogram(signal)
```

```python
filt_spec = madmom.audio.FilteredSpectrogram(spec)
```

```python
log_filt_spec = madmom.audio.LogarithmicSpectrogram(filt_spec)
```

```python
plt.imshow(log_filt_spec.T, origin='lower', aspect='auto')
plt.xlabel('frame')
for ann in annotations:
    plt.axvline(x=ann * FPS, color='w', linestyle=':', linewidth=2)
```

```python
log_filt_spec_flux = madmom.audio.SpectrogramDifference(log_filt_spec, positive_diffs=True)
spectral_flux = np.sum(log_filt_spec_flux, axis=1)
```

```python
act = np.load('/Users/matthewdavies/Desktop/cuidado_HideIt.npy')
```

```python
plt.plot(act)
plt.xlabel('frames')
plt.ylabel('activation')
for ann in annotations:
    plt.axvline(x=ann * FPS, color='r', linestyle=':', linewidth=2)
```

```python
plt.plot(spectral_flux)
plt.xlabel('frames')
plt.ylabel('activation')
for ann in annotations:
    plt.axvline(x=ann * FPS, color='r', linestyle=':', linewidth=2)
```
